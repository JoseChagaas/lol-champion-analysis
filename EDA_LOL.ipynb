```python
#Filtra a coluna client_positions apenas com valor "Top"
df_top = df[df['client_positions'].apply(lambda x: 'Top' in x)]
```

```python
#Filtra a coluna client_positions apenas com valor "Jungle"
df_jg = df[df['client_positions'].apply(lambda x: 'Jungle' in x)]
df_jg.head()
```

```python
#Filtra a coluna client_positions apenas com os valores 'Middle'
df_mid = df[df['client_positions'].apply(lambda x: 'Middle' in x)]
df_mid.head()
```

```python
#Função retorna True se 'Support' estiver presente no conjunto valor
def support(valor):
    return 'Support' in valor

#Aplica a função support em cada linha da coluna 'client_positions'
filter_sup = df['client_positions'].apply(support)

#Filtra o data frame com ajuda da variavel 'filter_sup'
df_sup = df[filter_sup]

df_sup.head()
```

```python
def adc(valor):
    return 'Bottom' in valor

filter_adc = df['client_positions'].apply(adc)

df_adc = df[filter_adc]
df_adc.head()
```

```python
#Pegando quantidade de campeão por data frame
qtd_top = len(df_top)
qtd_jg = len(df_jg)
qtd_mid = len(df_mid)
qtd_sup = len(df_sup)
qtd_adc = len(df_adc)
```

```python
print(qtd_top, qtd_jg, qtd_mid, qtd_sup, qtd_adc)
```

```python
lanes = ['Top', 'Jungle', 'Mid', 'Sup', 'ADC']
quantidades = [ qtd_top, qtd_jg, qtd_mid, qtd_sup, qtd_adc]

#Criando gráfico de barras QUANTIDADE X LANES
plt.figure(figsize=(8,5))
plt.bar(lanes, quantidades, color = 'skyblue')

#Títulos
plt.title('Campeões por Lane')
plt.xlabel('Lanes')
plt.ylabel('Quantidade')

plt.show()
```

```python
#Calculando a média de dificuldades dos campeões de cada lane
media_top = df_top['difficulty'].mean()
media_jg = df_jg['difficulty'].mean()
media_mid = df_mid['difficulty'].mean()
media_sup = df_sup['difficulty'].mean()
media_adc = df_adc['difficulty'].mean()

#Criando gráfico de barras DIFICULDADE x LANES

lanes = ['Top', 'Jungle', 'Mid', 'Sup', 'ADC']
dificuldade = [ media_top, media_jg, media_mid, media_sup, media_adc]

plt.figure(figsize=(8,5))
plt.bar(lanes, dificuldade, color='skyblue')

#Títulos
plt.title('Média da Dificuldade de Campeão por Lane (1 a 3)')
plt.xlabel('Lanes')
plt.ylabel('Dificuldade')

plt.show()
```
